# Analisis de datos de canales de youtube



# Preparacion de librerias y url

```python
import pandas as pd
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Canales_de_YouTube_con_m%C3%A1s_suscripciones'
```

### Uso del url y procesado

```python
list_youtube =  pd.read_html(url)
```

```python
list_youtube[0]
```

```python
df = list_youtube[0]
df.head()
```

```python
print(f"Shape: {df.shape}, Columns: {df.columns.tolist()}")
```

**Limpieza de columnas que no son necesarias**

```python
df = df.drop('N.º', axis=1)
df.head()
```

**Cambio de nombres de columnas**

```python
df.columns= ['Nombre', 'Pais', 'Idioma', 'Subscripciones (millones)']
df.head()
```

```python
df.at[0, 'Idioma'] = "Inglés"
df.head()
```

**Eliminacion de ultima fila**

```python
df.tail()
```

```python
df = df.drop([100])
df.tail()
```

```python
df.tail(20)
```
